```python
from collections import defaultdict

import nltk
import random
import numpy as np

from bs4 import BeautifulSoup
```

```python
positive_reviews = BeautifulSoup(open('../data/electronics/positive.review').read())
positive_reviews = positive_reviews.findAll('review_text')
```

```python
"""Create trigrams. What's the likelihood of WORD,i | WORD,i-1 and WORD,i+1
trigrams: {
    (WORD,i-1, WORD,i+1): [WORD,i, ...],
    ...
}
"""
trigrams = defaultdict(list)
for review in positive_reviews:
    # lowercase the review, don't want dupes
    s = review.text.lower()
    tokens = nltk.tokenize.word_tokenize(s)
    for i in range(len(tokens) - 2):
        trigrams[(tokens[i], tokens[i+2])].append(tokens[i+1])
```

```python
trigrams
```

```python
trigrams_probabilities = {}
for k, words in trigrams.items():
    # we only want to create probabilities for multiple words
    if len(set(words)) > 1:
        d = defaultdict(int)
        n = 0
        for w in words:
            d[w] += 1
            n += 1
        # convert to probs
        for w, c in d.items():
            d[w] = float(c) / n
            
        trigrams_probabilities[k] = d
```

```python
trigrams_probabilities
```

```python
def random_sample(d):
    r = random.random()
    cumulative = 0
    
    for w, p in d.items():
        cumulative += p
        if r < cumulative:
            return w
```

```python
def test_spinner():
    review = random.choice(positive_reviews)
    s = review.text.lower()
    print(f"Original: {s}")
    
    tokens = nltk.tokenize.word_tokenize(s)
    for i in range(len(tokens) - 2):
        if random.random() < 0.2:
            k = (tokens[i], tokens[i + 2])
            if k in trigrams_probabilities:
                w = random_sample(trigrams_probabilities[k])
                tokens[i + 1] = w
                
    print(f"Spun:")
    print({' '.join(tokens).replace(' .', '.').replace(' "', '"').replace(' ,', ',').replace(' $', '$').replace(' !', '!')})
```

```python
test_spinner()
```
